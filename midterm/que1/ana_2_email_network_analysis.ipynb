# Enron Data Anaysis : Email Network Analysis to find the number of emails sent and received per user 

```python
import nltk
import tarfile
import email
from os.path import splitext
import glob
import re
from os.path import basename
from email.parser import Parser
from collections import Counter
from operator import itemgetter
```

```python
files=glob.glob('C://Users//Jay Padhya//Downloads//maildir//*')
```

```python
# Analysis 1: Number of emailIds the employee has, mail received and sent by each Employee. 

employeeDetails=[]
number=0
for name in files:
    number+=1
    eName=basename(name)
    filesCategory=glob.glob(name+'/*')
    mailsSent=0
    mailsReceived=0
    employeeName=[]
    eEmailIds=[]
    if eName=='stokley-c':
        for chris in filesCategory:
            chrisFolder=glob.glob(chris+'/*')
            for emailCategory in chrisFolder:
                emailCategoryGeneral=emailCategory.split(chris+'\\')[1]
                if emailCategoryGeneral=='sent':
                    filesSent=glob.glob(emailCategory+'/*')
                    mailsSent=len(filesSent)
                    for eachEmail in filesSent:
                         with open(eachEmail) as data_file:
                            string_file=data_file.read()
                            email = Parser().parsestr(string_file)
                            eEmailIds.append(email['from'])
                            employeeName.append(re.sub("<.*?>","",email['X-From']))
    else:    
        for emailCategory in filesCategory:
            emailCategoryGeneral=emailCategory.split(name+'\\')[1]
            if emailCategoryGeneral=='sent':
                filesSent=glob.glob(emailCategory+'/*')
                mailsSent=len(filesSent)
                for eachEmail in filesSent:
                    with open(eachEmail) as data_file:
                        string_file=data_file.read()
                        email = Parser().parsestr(string_file)
                        eEmailIds.append(email['from'])
                        employeeName.append(re.sub("<.*?>","",email['X-From']))
            if emailCategoryGeneral=='inbox':
                filesInbox=glob.glob(emailCategory+'/*')
                mailsReceived=len(filesInbox)
                for inbx in filesInbox:
                    try:
                        with open(inbx) as data_file:
                            string_file=data_file.read()
                            email = Parser().parsestr(string_file)
#                             eEmailIds.append(email['To'])
                            employeeName.append(re.sub("<.*?>","",email['X-To']))
                    except:
                        pass
            if emailCategoryGeneral=='sent_items':
                filesSentItems=glob.glob(emailCategory+'/*')
                for sentItem in filesSentItems:
                    try:
                        with open(sentItem) as data_file:
                            string_file=data_file.read()
                            email = Parser().parsestr(string_file)
                            eEmailIds.append(email['from'])
                            employeeName.append(re.sub("<.*?>","",email['X-From']))
                    except:
                        pass
            if emailCategoryGeneral=='_sent_mail':
                filesSentMails=glob.glob(emailCategory+'/*')
                for sentMail in filesSentMails:
                    try:
                        with open(sentMail) as data_file:
                            string_file=data_file.read()
                            email = Parser().parsestr(string_file)
                            eEmailIds.append(email['from'])
                            employeeName.append(re.sub("<.*?>","",email['X-From']))
                    except:
                        pass       
    empName= Counter(employeeName).most_common(1)  
    noEmailIds=len(set(eEmailIds))        
    employeeDetails.append((number,empName[0][0],mailsSent,mailsReceived,noEmailIds))


#employeeDetails
```

```python
def get_text_from_email(msg):
    '''To get the content from email objects'''
    parts = []
    for part in msg.walk():
        if part.get_content_type() == 'text/plain':
            parts.append( part.get_payload() )
    return ''.join(parts)
```
