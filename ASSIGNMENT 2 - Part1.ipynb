```python
import glob,string #Importing glob for iterations and storing

filepath = 'gutenberg/shake*' #Relative path storing
# for name in glob.glob(filepath, recursive = True):
#     print (name) #fetches all word texts from Shakespreare- shakespeare-caesar.txt,shakespeare-hamlet.txt,shakespeare-macbeth.txt

for files in glob.glob(filepath,recursive = True): #fetching values
    data_list = open(files,"r").readlines() # Contains a text for all three works.

empty_datalist = []    
for x in data_list:
    empty_datalist.append(x.split())
data_list = [x for x in empty_datalist if x]
#data_list (To get the values of text)
```

```python
def lower_text(data_list): #The function to define all the lower words remove punctionation for the works 
    empty_list = [] #To store the lower words without punctuation
    punctuations = '''!()-[]{};:'"\,<>./?@#$%^&*_~''' #Check for punctuations
    for x in data_list: # Iterate over and remove punctuations first and then lowercase siultaneously
        for y in x:
            if y not in punctuations:
                empty_list.append(y.lower())
    return empty_list
        
shakespeare_lower_words = lower_text(data_list) #To return all the lower words without punctuation
#print(shakespeare_lower_words)#Prints all lower words and without punctuations 
```

```python
def makedictionary(data_list): #We are makig one more function for word count with lower text
    rs=[]
    for i in data_list:
        word_count_loweralpha = data_list.count(i) # Counting the words
        rs.append((i,word_count_loweralpha)) #Appednign to list
    return rs

shakespeare_dictionary = makedictionary(shakespeare_lower_words) # make a final dictionary
#print(shakespeare_dictionary) This prints the key value pair of dictionary
```

```python
dict_ = dict(shakespeare_dictionary) #COnverting to dict
dictionarySorted = sorted(dict_.items(),key = lambda kv:((-kv[1], kv[0]))) #To use a lambda function and sort correctly as keys
print(dictionarySorted) # To print the word count
```

```python
# Sorting in the format of Rank, Count, Word format:    
type(dictionarySorted)
counter = 1 #setting counter
ls_ten = [('WORD','COUNT','RANK')] # Making the format to word, count and rank successively

for x in dictionarySorted: # counter for the words
    ls_ten.append(x+(counter,)) #To add in the same Touple
    counter+=1

print(ls_ten) #To ensure we have rankwise sorting of the array    
```

```python
#Writing to CSV
import csv
with open('Zipf_output.csv','w') as fp: #Command to make csv
    a = csv.writer(fp, delimiter = ',',lineterminator='\n')
    a.writerows(ls_ten,) #writerows and writerow would give a single row
        
```
